```python
new_df = df.dropna()
new_df
```

```python
new_df = new_df.drop_duplicates()
new_df
```

```python
sorted_df = new_df.sort_values('Pop. Density (per sq. mi.)', ascending=False)
highest_density_country = sorted_df.iloc[0]['Country']
print("The country with the highest population density is:", highest_density_country)
```

```python
infant_mortality = new_df.sort_values('Infant mortality (per 1000 births)')
lowest_infant_mortality_country = infant_mortality.iloc[0]['Country']
print("The country with the lowest infant mortality rate is:", lowest_infant_mortality_country)
```

```python
gdp_per_capita = new_df.sort_values('GDP ($ per capita)', ascending=False)
highest_gdp_country = gdp_per_capita.iloc[0]['Country']
highest_gdp_value = gdp_per_capita.iloc[0]['GDP ($ per capita)']
print("The country with the highest GDP per capita is:", highest_gdp_country)
print("GDP per capita value:", highest_gdp_value)
```

```python
region_df = new_df[new_df['Region'] == 'Your Region']


try:
    # Get the literacy rate for the specific region
    literacy_rate = region_df['Literacy (%)'].values[0]

    # Print the result
    print("The literacy rate in Your Region is:", literacy_rate)
except IndexError:
    print("The specified region was not found in the data.")
```

```python
arable_land = new_df.sort_values('Arable (%)', ascending=False)
highest_arable_country = arable_land.iloc[0]['Country']
highest_arable_value = arable_land.iloc[0]['Arable (%)']
print("The country with the highest percentage of arable land is:", highest_arable_country)
print("Percentage of arable land:", highest_arable_value)

```
